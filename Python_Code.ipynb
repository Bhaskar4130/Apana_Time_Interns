```python
import pandas as pd
import numpy as np
```

```python
from google.colab import files
```

```python
uploaded_data = files.upload()
```

```python
data = list(uploaded_data.keys())[0]
df = pd.ExcelFile(data).parse('Sheet1')
```

```python
eda_results = {}
```

```python
a=df.describe().T
```

```python
numeric_df = df.select_dtypes(include=[np.number])
```

```python
print(numeric_df.fillna(value=0))
```

```python
summary = numeric_df.describe().T
```

```python
for column in df.select_dtypes(include=[np.number]).columns:
    column_data = df[column].fillna(value=0)  # Drop missing values
    eda_results[column] = {
        "Mean": column_data.mean(),
        "Median": column_data.median(),
        "Mode": column_data.mode()[0] if not column_data.mode().empty else None,
        "Variance": column_data.var(),
        "Standard Deviation": column_data.std(),
        "Range": column_data.max() - column_data.min(),
        "Skewness": column_data.skew(),
        "Kurtosis": column_data.kurt()
    }

# Convert results to a DataFrame for display
eda_df = pd.DataFrame(eda_results).T

# Display the EDA summary
print("Summary of EDA:")
print(eda_df)
```
